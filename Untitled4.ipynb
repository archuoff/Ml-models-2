```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= "http://bit.ly/w-data"
data1=pd.read_csv(data)
print(data1)
```

```python

data1.plot(x="Hours",y="Scores",style="o",color='y')
plt.title("prediction")
plt.xlabel("Hours")
plt.ylabel("Scores")
```

```python
X=data1.iloc[:,:-1].values
Y=data1.iloc[:,1].values
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
d=LinearRegression()
d.fit(x_train,y_train)
predictions=d.predict(x_test)
m=mean_squared_error(y_test,predictions)
plt.scatter(x_test,y_test,color="r",linewidth=0.3)
plt.plot(x_test,predictions,color="y",linestyle='--')
plt.title("hours vs percentage")
plt.xlabel("Hours")
plt.ylabel("Scores")
print(m)
```

```python
d1=pd.DataFrame({'Actual': y_test,"Predicted": predictions})
print(d1)
```
